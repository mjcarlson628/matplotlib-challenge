# Pymaceuticals Challenge

## Observations

 - Capomulin performed well compared to the other studied drugs for reducing tumor growth.
 - There is less variation in the final volume of tumors for mice treated with Capomulin than for the other three promising treatments.
 - Larger mice tended to grow larger tumors over the course of the study.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import sem
from scipy.stats import linregress
```

```python
# Mouse data and study results

mouse_metadata_df = pd.read_csv("Resources/Mouse_metadata.csv")
study_results_df = pd.read_csv("Resources/Study_results.csv")
```

```python
# Find any rows with identical Mouse ID and Timepoint values

duplicate_mice = study_results_df[study_results_df.duplicated(["Mouse ID", "Timepoint"], keep = False)]
duplicate_mice
```

```python
# All data associated with mouse g989

study_results_df.loc[study_results_df["Mouse ID"] == "g989"]
```

```python
# Remove all data associated with mouse g989

study_results_cleaned_df = study_results_df.drop(study_results_df.loc[study_results_df["Mouse ID"] == "g989"].index)
study_results_cleaned_df
```

```python
# Count the number of mice in the cleaned DataFrame

mice_total = len(study_results_cleaned_df["Mouse ID"].unique())
mice_total
```

```python
# Combine Mouse_metadata and Study_results into one DataFrame

combined_df = pd.merge(study_results_cleaned_df, mouse_metadata_df, on = "Mouse ID")
combined_df
```

## Summary Statistics

```python
# Tumor sizes based on drug regimen

# group by drug name and look at tumor sizes for each drug
tumor_stats = combined_df[["Drug Regimen", "Tumor Volume (mm3)"]].copy().groupby("Drug Regimen")["Tumor Volume (mm3)"]

# Calculate mean, median, standard deviation, and SEM
mean_vol = tumor_stats.mean()
median_vol = tumor_stats.median()
var_vol = tumor_stats.var()
stdev_vol = tumor_stats.std()
sem_vol = tumor_stats.sem()

# Summary statistics table
summary_table = pd.DataFrame({
    "Mean Volume (mm3)": mean_vol,
    "Median Volume (mm3)": median_vol,
    "Variance": var_vol,
    "Standard Deviation": stdev_vol,
    "SEM": sem_vol
})

summary_table
```

## Bar Charts

```python
# Organize the data to be able to compare the number of mice based on which drug they were given

drug_names = summary_table.index
mice_per_drug = combined_df.groupby("Drug Regimen")["Mouse ID"].value_counts().unique()

drug_counts_df = pd.DataFrame({
    "Drug Regimen": drug_names,
    "Mice": mice_per_drug
}).set_index(["Drug Regimen"])
```

## Quartiles, Outliers, and Boxplots

```python
# All data corresponding to Capomulin, Ramicane, Infubinol, and Ceftamin
treatments = ["Capomulin", "Ceftamin", "Infubinol", "Ramicane"]
treatment_list = combined_df.loc[combined_df["Drug Regimen"].isin(treatments)]

# Final timepoint for each mouse
max_time = treatment_list.groupby("Mouse ID")["Timepoint"].max()

# left join to collect only the final time for each mouse on the four drug regimens
final_measurements_df = pd.merge(max_time, treatment_list, on = ["Mouse ID", "Timepoint"], how = "left")
final_measurements_df
```

```python
# Lists of final tumor sizes of each drug regimen

capomulin_list = final_measurements_df.loc[final_measurements_df["Drug Regimen"] == "Capomulin"]["Tumor Volume (mm3)"]
ceftamin_list = final_measurements_df.loc[final_measurements_df["Drug Regimen"] == "Ceftamin"]["Tumor Volume (mm3)"]
infubinol_list = final_measurements_df.loc[final_measurements_df["Drug Regimen"] == "Infubinol"]["Tumor Volume (mm3)"]
ramicane_list = final_measurements_df.loc[final_measurements_df["Drug Regimen"] == "Ramicane"]["Tumor Volume (mm3)"]
```

```python
# DataFrame of quartiles for final tumor size with each drug

quartiles = pd.DataFrame({
    "Lower Quartile": final_measurements_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].quantile(0.25),
    "Median": final_measurements_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].quantile(0.5),
    "Upper Quartile": final_measurements_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].quantile(0.75),
})

quartiles
```

```python
# interquartile ranges for each drug regimen

iqr_dict = {}
for x in treatments:
    iqr = (quartiles.loc[x]["Upper Quartile"] - quartiles.loc[x]["Lower Quartile"])
    iqr_dict[x] = iqr

# calculate min and max tumor volumes to check for potential outliers
    min_size = final_measurements_df.loc[final_measurements_df["Drug Regimen"] == x]["Tumor Volume (mm3)"].min()
    max_size = final_measurements_df.loc[final_measurements_df["Drug Regimen"] == x]["Tumor Volume (mm3)"].max()
    
    lower_bound = quartiles.loc[x]["Lower Quartile"] - 1.5 * iqr_dict[x]
    upper_bound = quartiles.loc[x]["Upper Quartile"] + 1.5 * iqr_dict[x]
    print(f"{x}: Values below {lower_bound} or above {upper_bound} could be outliers.")
    if min_size < lower_bound or max_size > upper_bound:
        print("-----------------------------------------")
        print(f"{x} has a potential outlier.")
        print("-----------------------------------------")        
```

```python
# DataFrame of IQR values for each drug regimen
iqr_df = pd.DataFrame([iqr_dict]).rename(index = {0: "IQR"})
iqr_df
```

```python
data = [capomulin_list, ceftamin_list, infubinol_list, ramicane_list]
x_labels = treatments
fig1, ax1 = plt.subplots()

# Display potential outliers as red squares
ax1.boxplot(data, flierprops = {"markerfacecolor": "r", "marker": "s"})

ax1.set_xlabel("Drug Regimen")
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.set_xticklabels(x_labels)
ax1.set_title("Final Tumor Volume for each Drug Regimen")

plt.show()
```

## Scatter Plot

```python
# Average tumor volume for all Capomulin data points

avg_volume = combined_df.loc[combined_df["Drug Regimen"] == "Capomulin"]["Tumor Volume (mm3)"].mean()
avg_volume
```

```python
# Check to see if any mouse's weight changed throughout the trial by comparing max and min weight for each mouse
# All mice stayed at the same weight for all measurements

combined_df.loc[combined_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID")["Weight (g)"].max() == combined_df.loc[combined_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID")["Weight (g)"].min()
```

```python
# Calculate the weight and average tumor volume for each Capomulin mouse

mouse_weight = combined_df.loc[combined_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID")["Weight (g)"].mean()
avg_tumor = combined_df.loc[combined_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID")["Tumor Volume (mm3)"].mean()


pd.DataFrame({
    "Weight (g)": mouse_weight,
    "Average Tumor Volume": avg_tumor
    
})
```

```python
# Scatter Plot of mouse weight vs average tumor volume
x_values = mouse_weight
y_values = avg_tumor

plt.figure(figsize = (8, 8))
plt.scatter(x_values, y_values)
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor volume (mm3)")
plt.title("Average Tumor Volume vs Mouse Weight")

plt.show()
```

## Correlation and Regression

```python
# Find the linear regression line for tumor volume vs mouse weight
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
plt.scatter(x_values,y_values)

# Show the regression line
plt.plot(x_values,regress_values,"r-")

# Write the equation and correlation coefficient on the plot
plt.annotate(f"y = {round(slope,2)}x + {round(intercept,2)}", (20, 37), fontsize=15,color="red")
plt.annotate(f"r = {round(rvalue, 2)}", (20, 35), fontsize=15,color="red")

plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor volume (mm3)")
plt.title("Average Tumor Volume vs Mouse Weight")

plt.show()
```
